# Unsupervised Learning- Credit Card fraud Detection

```python
#Importing required libraries
import numpy as np                       #matrix form
import matplotlib.pyplot as plt          #for graphing visualization
import pandas as pd
```

```python
#loading dataset
df = pd.read_csv('creditcard.csv', low_memory=False)
df.head()
```

```python
from sklearn.cluster import KMeans  #KMeans algorithm
from sklearn import metrics         #accuracy calculation
from sklearn.decomposition import PCA  #reducing dimensionality
from sklearn.preprocessing import scale #for scaling
from sklearn.model_selection import train_test_split #splitting data into training and testing data
```

```python
X = df.iloc[:,:-1] #defining dependent variable y and independent variables in X 
y = df['Class']

X_scaled = scale(X) #scaling for converting to 2d array 
pca = PCA(n_components=2) #reducing dimensionality using PCA
X_reduced = pca.fit_transform(X_scaled) #fitting the scaled data according to the dimensions

#splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_reduced, y, test_size = 0.33, random_state=500)
```

```python
#our model "kmeans" is built using KMeans method with n_clusters that is 2 clusters , init parameter being method of initialization to be one incrementing for k
kmeans = KMeans(init='k-means++', n_clusters=2, n_init=10)#n_init is for total number of times the k-means algo should run for different centroid seeds
kmeans.fit(X_train) #fitting our model
```

```python
# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .01     # point in the mesh [x_min, x_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = X_reduced[:, 0].min() - 1, X_reduced[:, 0].max() + 1
y_min, y_max = X_reduced[:, 1].min() - 1, X_reduced[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
#Plotting the data-visualization
# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.plot(X_reduced[:, 0], X_reduced[:, 1], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on the credit card fraud dataset (PCA-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
#predicting for testing data
predictions = kmeans.predict(X_test)
#defining the variables for comparison of predicted and actual values
pred_fraud = np.where(predictions == 1)[0]#predicted value for fraud
real_fraud = np.where(y_test == 1)[0]#actual value for fraud
false_pos = len(np.setdiff1d(pred_fraud, real_fraud))#checking the correct predictions for a fraud

pred_good = np.where(predictions == 0)[0]#predicted value for no fraud 
real_good = np.where(y_test == 0)[0]#actual value for no fraud  
false_neg = len(np.setdiff1d(pred_good, real_good))##checking the correct predictions for a fraud

false_neg_rate = false_neg/(false_pos+false_neg)#determining the error

accuracy = (len(X_test) - (false_neg + false_pos)) / len(X_test)#calculating accuracy
print("Accuracy:", accuracy)
print("False negative rate (with respect to misclassifications): ", false_neg_rate)
print("False negative rate (with respect to all the data): ", false_neg / len(predictions))
print("False negatives, false positives, mispredictions:", false_neg, false_pos, false_neg + false_pos)
print("Total test data points:", len(X_test))
```
